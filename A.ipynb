```python
import numpy as np
import pandas as pd

PATH: str = "data/ratings.dat"

USER_CNT, MOVIE_CNT = (6040, 3952)
data = np.ndarray((USER_CNT, MOVIE_CNT))

with open(PATH, "r") as f:
    lines = f.readlines()
    for line in lines:
        user_id, movie_id, rating, _ = map(int, line.split("::"))
        data[user_id - 1][movie_id - 1] = rating

df = pd.DataFrame(data)
df
```

```python
from sklearn.cluster import KMeans

n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
user_groups = kmeans.fit_predict(df)
df[2]
```

```python
group = [[] for _ in range(n_clusters)]
for i in range(USER_CNT):
    group[user_groups[i]].append(df.iloc[i])

group_counts = [len(g) for g in group]
group_counts_df = pd.DataFrame({'Group': np.arange(1, len(group_counts) + 1), 'Number of Users': group_counts})

group_counts_df
```

```python
for i in range(n_clusters):
    group_df = pd.DataFrame(group[i])
    print(f"Group {i}:\n{group_df}\n")
```

```python
for i in range(n_clusters):
    group_df = pd.DataFrame(group[i])
    res = pd.DataFrame()

    res['AU'] = group_df.apply(lambda t: t.sum())
    res['AVG'] = group_df.apply(lambda t: t.mean())
    res['SC'] = group_df.apply(lambda t: t[t > 0].sum())
    res['AV'] = group_df.apply(lambda t: t[t >= 4].sum())
    res['BC'] = group_df.apply(lambda t: (t.rank(axis=0, method='average')).sum())

    print(f"Group {i}:\n{res}\n")
```
